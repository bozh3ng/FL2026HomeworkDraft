## Exercise 1 

Ridge Regression vs. Ordinary Least Squares (25 pts)

Goal: Empirically illustrate regularization.
1. Data - Generate a synthetic regression dataset with `make_regression` (100 samples, 50 features, SNR $\approx 10 \mathrm{~dB}$ ). Keep $20 \%$ for validation.
2. Models - Implement
- OLS ( $\alpha=0$ )
- Ridge for $\alpha \in\{1 \mathrm{e}-4,1 \mathrm{e}-3, \ldots, 1 \mathrm{e} 3\}$.
3. Plots
- Training vs. validation MSE as a function of $\log_{10}\alpha$
- $\|\mathrm{w}\|_2$ vs. $\alpha$
4. Report - Identify the $a$ that minimizes validation error and explain in $\leq 100$ words how the curve demonstrates the bias-variance trade-off.

Hints: Use `sklearn.linear_model.Ridge`, `StandardScaler`; set `random_state=42` for reproducibility.

```python
# fl_hw1_solutions.py  –– Homework #1  (Foundations refresher)
# Reproducibility
import numpy as np, torch, random, matplotlib.pyplot as plt
random_state = 42
np.random.seed(random_state); random.seed(random_state); torch.manual_seed(random_state)

# =============================================================================
# Exercise 1  –– Ridge Regression vs. Ordinary Least Squares
# =============================================================================
from sklearn.datasets import make_regression
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error

# 1. Synthetic data -----------------------------------------------------------
X, y = make_regression(
    n_samples=100, n_features=50, noise=15.0, random_state=random_state
)
idx = np.random.permutation(len(X))
train_idx, val_idx = idx[:80], idx[80:]
X_train, X_val = X[train_idx], X[val_idx]
y_train, y_val = y[train_idx], y[val_idx]

# 2–3. Model sweep + plots -----------------------------------------------------
alphas = np.logspace(-4, 3, num=8)      # 1e-4 … 1e3
train_mse, val_mse, w_norm = [], [], []

for a in alphas:
    model = make_pipeline(StandardScaler(), Ridge(alpha=a, random_state=random_state))
    model.fit(X_train, y_train)
    y_tr_hat = model.predict(X_train)
    y_val_hat = model.predict(X_val)

    train_mse.append(mean_squared_error(y_train, y_tr_hat))
    val_mse.append(mean_squared_error(y_val, y_val_hat))
    # coefficient vector lives in model['ridge'].coef_
    w_norm.append(np.linalg.norm(model[-1].coef_))

best_a = alphas[int(np.argmin(val_mse))]
print(f"Optimal $\alpha$ = {best_a:.1e}")

# --- Plot (1): MSE vs log10 α -------------------------------------------------
plt.figure()
plt.semilogx(alphas, train_mse, marker='o', label='Train MSE')
plt.semilogx(alphas, val_mse, marker='s', label='Validation MSE')
plt.axvline(best_a, ls='--', lw=1)
plt.xlabel(r'$\alpha$')
plt.ylabel('MSE')
plt.legend(); plt.title('Bias-Variance trade-off (MSE)')
plt.show()

# --- Plot (2): ‖w‖₂ vs α ------------------------------------------------------
plt.figure()
plt.semilogx(alphas, w_norm, marker='d')
plt.xlabel(r'$\alpha$'); plt.ylabel(r'$\,\|w\|_2$')
plt.title('Weight norm shrinks with regularization')
plt.show()
```

## Exercise 2 

Parametric vs. Non-Parametric Classification (20 pts)

Goal: Compare logistic regression (parametric) with kNN (non-parametric) on a lowdimensional task.

1. Data - Load the two-moons dataset ( `make_moons`, noise $=0.25$ ). Split $70 / 30$.
2. Grid Search -
- Logistic regression with $C \in\{0.01,0.1,1,10\}$.
- kNN with $\mathrm{k} \in\{1,3,5,15,30\}$.
3. Metrics - Accuracy + decision boundary plots.
4. Discussion - In $\leq 150$ words compare decision boundaries, dependence on trainingset size, and memory footprint.

```python
# =============================================================================
# Exercise 2  –– Parametric (LogReg) vs. Non-Parametric (kNN) Classification
# =============================================================================
from sklearn.datasets import make_moons
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. Data ---------------------------------------------------------------------
Xm, ym = make_moons(n_samples=1000, noise=0.25, random_state=random_state)
Xm_tr, Xm_te, ym_tr, ym_te = train_test_split(
    Xm, ym, test_size=0.3, random_state=random_state, stratify=ym
)

# 2. Grid search --------------------------------------------------------------
Cs   = [0.01, 0.1, 1, 10]
ks   = [1, 3, 5, 15, 30]
accL, accK = {}, {}

for C in Cs:
    clf = LogisticRegression(C=C, solver='lbfgs')
    clf.fit(Xm_tr, ym_tr)
    accL[C] = accuracy_score(ym_te, clf.predict(Xm_te))

for k in ks:
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(Xm_tr, ym_tr)
    accK[k] = accuracy_score(ym_te, clf.predict(Xm_te))

best_C = max(accL, key=accL.get)
best_k = max(accK, key=accK.get)

print("LogReg accuracies:", accL)
print("kNN   accuracies:", accK)
print(f"Best LogReg: C={best_C}  acc={accL[best_C]:.3f}")
print(f"Best kNN   : k={best_k}  acc={accK[best_k]:.3f}")

# --- Decision boundary visual -------------------------------------------------
def plot_decision_boundary(clf, title):
    x_min, x_max = Xm[:,0].min()-0.5, Xm[:,0].max()+0.5
    y_min, y_max = Xm[:,1].min()-0.5, Xm[:,1].max()+0.5
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 300), np.linspace(y_min, y_max, 300)
    )
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    plt.figure(figsize=(4,3))
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(Xm_te[:,0], Xm_te[:,1], c=ym_te, s=15, edgecolor='k')
    plt.title(title); plt.xticks([]); plt.yticks([]); plt.show()

plot_decision_boundary(
    LogisticRegression(C=best_C, solver='lbfgs').fit(Xm_tr, ym_tr),
    f"Logistic Regression (C={best_C})"
)
plot_decision_boundary(
    KNeighborsClassifier(n_neighbors=best_k).fit(Xm_tr, ym_tr),
    f"kNN (k={best_k})"
)
```

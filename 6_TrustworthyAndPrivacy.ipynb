# Quiz 

Recall 

Definition 1. A stochastic algorithm $\mathcal{A}$ is $(\varepsilon, \delta)-D P$ if, for any two neighbouring datasets $\mathcal{D}, \mathcal{D}^{\prime}$,

$$
\operatorname{Prob}\{\mathcal{A}(\mathcal{D}) \in \mathcal{S}\} \leq \exp (\varepsilon) \operatorname{Prob}\left\{\mathcal{A}\left(\mathcal{D}^{\prime}\right) \in \mathcal{S}\right\}+\delta
$$

holds for every measurable set $\mathcal{S}$.


Definition 2.  An algorithm $\mathcal{A}$ is $(\alpha, \gamma)$-RDP if, for any two neighbouring datasets $\mathcal{D}$ and $\mathcal{D}^{\prime}$,

$$
D_\alpha\left(p(\widehat{\mathbf{w}} ; \mathcal{D}) \| p\left(\widehat{\mathbf{w}} ; \mathcal{D}^{\prime}\right)\right) \leq \gamma
$$
